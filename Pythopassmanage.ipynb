# Password manager

```python
import pandas as pd
import os 
import csv

file_name = "passwordmanagent.csv"

def load_accounts():
    records = [] 
    if not os.path.exists(file_name):
        return records
    try:
        with open(file_name, "r") as f:
            reader = csv.DictReader(f)
            for row in reader:
                records.append(row)
    except Exception as e:
        print("Error found:", e)

    return records
        
    
def save_account(records):
    try:
        with open(file_name, "w", newline="") as f:
            writer = csv.DictWriter(f, fieldnames=["website", "username", "password"])
            writer.writeheader()
            writer.writerows(records)
    except Exception as e:
        print("Error found:", e)

def add_account():
    website = input("Website : ").strip().lower()
    username = input ("Username :").strip()
    password = input("Password :").strip()

    if not website or not username or not password:
        print("All fields required.")
        return
    
    records = load_accounts()
    records.append({"website": website,"username": username,"password": password})
    save_account(records)
    print(F"Account info for {website} saved g")
    
def retrieve_account():
    website = input("which site info you looking for").strip().lower()
    matches = [r for r in load_accounts() if r["website"] == website]
    
    if not matches:
        print(f"No account with such name")
        return
    
    print(f"\n {website}")
    for i, r in enumerate(matches,1):
        print(f" {i}.Username : {r['username']}")
        print(f" Password : {r['password']}")

def delete_account():
    website = input("You are deleting info, select website")
    username = input("you are deleting data,chose username")

    records = load_accounts()
    updated = [r for r in records if not (r["website"] == website and r["username"] == username)]

    if len(updated) == len(records):
        print(f"No match found ")
        return
    confirm = input(f"Delete'{username}' from '{website}'?.").strip().lower()
    if confirm !="yes":
        print("Deletion cancelled")

    save_account(updated)
    print(f"Account deleted")

def list_accounts():
    records = load_accounts()
    if not records:
        print("No account info stored")
        return
    print ("\n stored account info")
    for i, r in enumerate(sorted(records,key=lambda x: x["website"]),1):
        print(f"{i}.{r['website']} - {r['username']}") 

while True:
    print("\n Choose service(1-5)")
    print("1.Add account")
    print("2.retrieve account")
    print("3.Delete account")
    print("4.list accounts")
    print("5.Exit")

    choice = input("choose service:")

    if choice == "1":
        add_account()

    elif choice =="2":
        retrieve_account()

    elif choice =="3":
        delete_account()
        print(f"youve deleted {'website'} acc info")

    elif choice == "4":
        list_accounts()

    elif choice == "5":
        print("see yaa")
        break        












    






    
    


    



          


        
```
